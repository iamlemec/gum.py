```python
G.Box(
    G.HStack(
        G.Square(rounded=True, fill=C.blue),
        G.Square(rounded=True, fill=C.red),
        spacing=0.05
    ),
    margin=True
)
```

```python
G.Frame(G.Text('Hello, world!'), rounded=True, padding=True, margin=True)
```

```python
path1 = G.SymLine(fy=C.sin, xlim=(0, 2*C.pi), stroke=C.blue, stroke_width=2)
path2 = G.SymLine(fy=C.cos, xlim=(0, 2*C.pi), stroke=C.red , stroke_width=2)
G.Box(
    G.VStack(
        G.Plot(path1, ylim=(-1.5, 1.5), aspect=2, grid=True),
        G.Plot(path2, ylim=(-1.5, 1.5), aspect=2, grid=True),
        spacing=True
    ),
    margin=True
)
```
