```python
import gum
import random
import pandas as pd
from gum import G, C
```

```python
G.Box(
    G.HStack(
        G.Square(rounded=True, fill=C.blue),
        G.Square(rounded=True, fill=C.red),
        spacing=0.05
    ),
    margin=True
)
```

```python
G.Frame(G.Text('Hello, world!'), rounded=True, padding=True, margin=True)
```

```python
path1 = G.DataPath(fy=C.sin, xlim=(0, 2*C.pi), stroke=C.blue, stroke_width=2)
path2 = G.DataPath(fy=C.cos, xlim=(0, 2*C.pi), stroke=C.red , stroke_width=2)
G.Box(
    G.VStack(
        G.Plot(path1, ylim=(-1.5, 1.5), aspect=2, grid=True),
        G.Plot(path2, ylim=(-1.5, 1.5), aspect=2, grid=True),
        spacing=True
    ),
    margin=True
)
```

```python
data = gum.test_data('brown', T=500)
gum.lines(data, ylim=(-2, 2), show=False)
```

```python
data = pd.Series({ chr(65+x): random.randint(1, 9) for x in range(20) })
gum.bars(data, ylim=(0, 10), show=False)
```
